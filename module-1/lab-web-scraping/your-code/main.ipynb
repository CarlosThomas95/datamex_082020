####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url5 = 'https://www.emsc-csem.org/Earthquake/'
resp5 = req.get(url5)
sopa5 = bs(resp5.content,'html.parser')
titles=['Date & Time UTC', 'Latitude degrees','Longitude degrees','Region name']
sismi=sopa5.find_all('td',{'class':'tabev6'})
lat=sopa5.findAll('td',{'class':'tabev1'})
lat1=sopa5.findAll('td',{'class':'tabev2'})
reg=sopa5.find_all('td',{'class':'tb_region'})
tot=[]
````

````python
for j in range(20):
    horas=(sismi[j].text.replace('earthquake','').replace('\xa0\xa0\xa0',' '))
    lati=(lat[2*j+1].text.replace('\xa0','')+' '+lat1[3*j+1].text.replace('\xa0\xa0',''))
    long=(lat[2*j].text.replace('\xa0','')+' '+lat1[3*j].text.replace('\xa0\xa0',''))
    zona=(reg[j].text.replace('\xa0',''))
    tot.append([horas,lati,long,zona])
````

````python
pd.DataFrame(tot, columns=titles)
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url5 = 'https://www.wikipedia.org/'
resp5 = req.get(url5)
sopa5 = bs(resp5.content,'html.parser')
````

````python
import re
lang=sopa5.find('div',{'class':'central-featured'}).find_all('div')
print(lang[0].text)
languages=[]
for k in lang:
    lan=re.findall(r'\w+',k.text)[0]
    art=''.join(re.findall(r'\d+',k.text))
    languages.append([lan, art])
languages
````

#### IMDB's Top 250 data (movie name, Initial release, director name and stars) as a pandas dataframe

````python
# This is the url you will scrape in this exercise 
url8 = 'https://www.imdb.com/chart/top'
````

````python
resp8 = req.get(url8)
sopa8 = bs(resp8.content,'html.parser')
````

````python
peliculas=[]
films=sopa8.find('tbody',{'class':'lister-list'}).find_all('td')
for l in range(249):
    name=re.findall(r'\w+ ?\w+ ?\w+ ?\w+ ?\w ?\w',films[5*l+1].text)#[0] 
    date=re.findall(r'\(\d+',films[5*l+1].text)[0].replace('(','')
    stars=films[5*l+2].text.replace('\n','')
    peliculas.append([name, date, stars])

peliculas

````
