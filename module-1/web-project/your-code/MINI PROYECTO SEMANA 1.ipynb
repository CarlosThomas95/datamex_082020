```python
import random    
frutas = ["cereza","carambola","platano","mango","manzana","durazno","melon","sandia","papaya","coco"]
carros = ["peugeot","nissan","toyota","honda","ford","dodge","chevrolet,","subaru","gmc","acura"]
animales = ["perro","gato","pez","leon","tigre","pato","elefante","tiburon","vibora","murcielago"]
plantas = ["alcornoque","almendro","berenjena","cactus","campanilla","drosera","eucalipto","francesilla","gordolobo","jacaranda"]
nombres = ["carlos","katia","gilberto","juana","yoshira","francisca","catalina","reynalda","alejadro","hector"]

def lista_play():
    x = random.randint(1,4)
    if x == 1:
        lista_player = frutas
        frutas.append("Frutas")
    elif x== 2:
        lista_player = carros
        carros.append("Carros")
    elif x== 3: 
        lista_player = animales
        animales.append("Animales")
    else:
        lista_player = nombres
        nombres.append("Nombres propios")
    return (lista_player)


print("Bienvenido al el AHORCADO, ingresa nombre de jugador:")
name = input()
print("**************************************************************************************************************************")
print(name, ", ahora debes adivinar la palabra escondida, para ellos debes ingresar letra por letra. Tienes tres oportunidades")
palabra = lista_play()
print ("Pista: Tu palabra pertence a una lista de: " ,palabra[-1])


x = random.randint(0,9)
palabra_adivinar = palabra[x]
print(palabra_adivinar)

ahorcado = []
for a in range(len(palabra_adivinar)):
    ahorcado.append("_")
print("Palabra: ",ahorcado)

aciertos = len(palabra_adivinar)
errores = 3
error_lista=[".",".","."]

while aciertos>0 and errores>0:
    print("***************************************************************************************")
    print("Ingrese una letra del abecedario, la palabra a adivinar puede o no incluir caracteres: ")
    word_player = input()
    word_player = word_player.lower()
    

    if word_player in palabra_adivinar:
        aciertos -= palabra_adivinar.count(word_player)
        for i in range(len(palabra_adivinar)):
            if word_player == palabra_adivinar[i]:
                x=word_player
                ahorcado[i]=x
        print("Palabra: ",ahorcado,"*****CORRECTO*****")
        print("Intentos fallidos: ",error_lista)
    else:
        errores -= 1
        print("Palabra",ahorcado,"*****FALLASTE*****")
        if error_lista[0]=="X":
            if error_lista[1]=="X":
                if error_lista[2]=="X":
                    pass
                else:
                    error_lista[2]="X"
            else:
                error_lista[1]="X"
        else:
            error_lista[0]="X"

        print("Intentos fallidos: ",error_lista)
if aciertos==0:
    print("---------------------------------------------------")
    print("Felicidades ",name," GANASTE!!!")
elif errores==0:
    print("---------------------------------------------------")
    print("Suerte para la proxima, sigue intentandolo",name,".")
           
    
```
